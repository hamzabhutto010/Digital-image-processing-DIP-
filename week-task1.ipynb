```python
# Use numpy to create a blank grayscale image
# (an array of zeros) with height=200 and width=300.
# give value to each pixel is 180
img = np.zeros((200,300),dtype='uint8')
print(img.shape)

for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        img[i,j] = 180
# img = img + 180
        
plt.imshow(img , cmap='gray' , vmin=0, vmax=255)
```

```python
# Create an image of shape 200x256 using numpy as shown below
img = np.zeros((200,256))

for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        img[i,j] = i

plt.imshow(img,cmap='gray', vmin=0 , vmax=256)
```

```python

img = cv2.imread('rose.tif')
img2 = np.random.rand(10,10)

level = 16

quantized = np.round(img / (256/level) * (256/level))

quantized = quantized.astype(np.uint8)

fig, axe = plt.subplots(2,2, figsize = (10,7))

axe[0,0].imshow(img)
axe[0,0].axis('off')
axe[0,1].imshow(quantized)
axe[0,1].axis('off')
axe[1,0].imshow(img2, cmap='gray')
axe[1,0].axis('off')

plt.axis('off')
```

```python
def gray_level(level):
    
    img = cv2.imread('rose.tif')
    img2 = np.random.rand(10,10)

    quantized = np.round((img // level ) * level)

    quantized = quantized.astype(np.uint8)

    fig, axe = plt.subplots(2,2, figsize = (7,7))

    axe[0,0].imshow(img)
    axe[0,0].axis('off')
    axe[0,1].imshow(quantized)
    axe[0,1].axis('off')
    axe[1,0].imshow(img2, cmap='gray')
    axe[1,0].axis('off')

    plt.axis('off')
    
gray_level(32)
    
```
