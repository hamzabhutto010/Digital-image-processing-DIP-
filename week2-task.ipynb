```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

img = cv2.imread('images/rose.tif')
img1 = cv2.imread('images/rose.tif')

mask = (img >= 100)
mask2 = (img >= 40) & (img <= 42)  

r, c, ch = np.where(mask2)
rows, colums, ch= np.where(mask)

img[rows,colums,2] = 255
img[r,c,1] = 255
img = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.axis('off')
plt.title("rose")
```

```python
# downsampling the image 
img = cv2.imread('images/rose.tif')

img512 = img[::2,::2]
img256 = img512[::2,::2]
img128 = img256[::2,::2]
img64 = img128[::2,::2]

fig , axe = plt.subplots(2,2, figsize = (7,7))

axe[0,0].imshow(img512)
axe[0,0].axis('off')
axe[0,0].set_title('512 x 512')

axe[0,1].imshow(img256)
axe[0,1].axis('off')
axe[0,1].set_title('256 x 256')

axe[1,0].imshow(img128)
axe[1,0].axis('off')
axe[1,0].set_title('128 x 128')

axe[1,1].imshow(img64)
axe[1,1].axis('off')
axe[1,1].set_title('64 x 64')

plt.tight_layout()
```

```python
# upsampling of image 64 x 64 to 128 x 128
ups = np.zeros((128 , 128, 3), dtype=('uint8'))

ups[::2, 0::2] = img64

plt.imshow(ups)
ups.shape
```

```python
# upsampling image using the interpolation 
# img64
upsmpling = np.zeros((512,512,3), dtype=('uint8'))

rows_scale = 64 / 512
cols_scale = 64 / 512

for i in range(upsmpling.shape[0] - 1):
    r = int(i * rows_scale)
    for j in range(upsmpling.shape[1] - 1):
        c = int(j * cols_scale)
        upsmpling[i,j] = img64[r,c]

fig , axe = plt.subplots(1,2, figsize=(7,7) )

axe[0].imshow(img64)
axe[1].imshow(upsmpling)
        
```

```python
# image quantizer from 8 bit - 7 - 6 - 4
img = cv2.imread('images/einstein.tif')

img7 = (img / 2).astype('uint8') * 2
img6 = (img / 4).astype('uint8') * 4
img5 = (img / 8).astype('uint8') * 8
img4 = (img / 16).astype('uint8') * 16
img3 = (img / 32).astype('uint8') * 32
img2 = (img / 64).astype('uint8') * 64
img1 = (img / 128).astype('uint8') * 255

fig, axe = plt.subplots(2,4, figsize = (7,7))

axe[0,0].imshow(img)
axe[0,0].axis('off')

axe[0,1].imshow(img7)
axe[0,1].axis('off')

axe[0,2].imshow(img6)
axe[0,2].axis("off")

axe[0,3].imshow(img5)
axe[0,3].axis('off')

axe[1,0].imshow(img4)
axe[1,0].axis('off')

axe[1,1].imshow(img3)
axe[1,1].axis('off')

axe[1,2].imshow(img2)
axe[1,2].axis('off')

axe[1,3].imshow(img1)
axe[1,3].axis("off")

plt.tight_layout()
```
