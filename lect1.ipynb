```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import pandas as pd

img = cv2.imread('einstein.tif',0)
print(img.shape)
result = np.zeros((img.shape[0]//2,img.shape[1]//2), dtype='float32')

for i in range(img.shape[0] - 1):
    for j in range(img.shape[1]-1):
        result[i//2,j//2] = img[i,j]
        
result = result.astype('int32')
print(result.shape)
plt.axis('off')
plt.imshow(result,cmap='gray')

```

```python
# image slicing 
image_panel = np.zeros((200,200,3), dtype = np.uint8)

# select row 0 - 99 and col 0 - 99
image_panel[:100,:100,0] = 255
# select row 0 - 99 col 100 - last
image_panel[:100,100:,1] = 255
# slect row 99 - last col 0 - 100
image_panel[100:,:,2] = 255

plt.imshow(image_panel)
plt.axis('off')
plt.show()
```

```python
# quantization 
img = cv2.imread('einstein.tif',0)

result_7_bit = (img // 2).astype('uint8') * 2
result_6_bit = (img // 4).astype('uint8') * 4
result_5_bit = (img // 8).astype('uint8') * 8
result_4_bit = (img // 16).astype('uint8') * 16
result_3_bit = (img // 32).astype('uint8') * 32
result_3_bit = (img // 64).astype('uint8') * 64
result_2_bit = (img // 128).astype('uint') * 128

plt.imshow(result_7_bit, cmap='gray')
plt.imshow(result_2_bit, cmap='gray')


```

```python
img = cv2.imread('rose.tif', 0)

img512 = img[::2, ::2];
img256 = img512[::2, ::2];
img128 = img256[::2, ::2];
img64 = img128[::2, ::2];
img32 = img64[::2, ::2];
            
# create figure
fig, axs = plt.subplots(2, 3, figsize=(10, 7))
  
# showing images
axs[0,0].imshow(img,cmap = 'gray')
axs[0,0].axis('off')
axs[0,0].set_title('Original image')

axs[0,1].imshow(img512, cmap = 'gray')
axs[0,1].axis('off')
axs[0,1].set_title('result512')

axs[0,2].imshow(img256,cmap = 'gray')
axs[0,2].axis('off')
axs[0,2].set_title('result256')

axs[1,0].imshow(img128,cmap = 'gray')
axs[1,0].axis('off')
axs[1,0].set_title('result128')

axs[1,1].imshow(img64,cmap = 'gray')
axs[1,1].axis('off')
axs[1,1].set_title('result64')

axs[1,2].imshow(img32,cmap = 'gray')
axs[1,2].axis('off')
axs[1,2].set_title('result32')

plt.tight_layout();
plt.show()
```
