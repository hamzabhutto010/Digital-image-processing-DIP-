```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
from matplotlib import figure
from numpy import median


img = cv2.imread('images/circuit.tif')
# using 7 x 7 kernel 
m = 3
n = 3

p = (m - 1)//2

# padded_img = np.pad(img, ((p_m,p_m), (p_n,p_n)), mode='constant', constant_values=0)
# padded_img = np.pad(img, ((p_m,p_m), (p_n,p_n)), mode='constant', constant_values=0)
padded_img = np.pad(img, ((p, p), (p, p), (0, 0)), mode='constant', constant_values=0)

output_img = np.zeros_like(img)

rows = img.shape[0]
cols = img.shape[1]

for row in range(rows):
    for col in range(cols):
        reg = padded_img[row:row + m, col: col + n]
        oneD = np.ravel(reg)
        sort_val = np.sort(oneD)
        median = np.median(sort_val)
        output_img[row, col] = median


fig , axe = plt.subplots(1,2, figsize = (10,5))
axe[0].imshow(img, cmap='gray')
axe[0].set_title('original x ray image')

axe[1].imshow(output_img, cmap='gray')
axe[1].set_title("median filter apply")
```
